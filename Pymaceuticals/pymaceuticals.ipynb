## Observations and Insights 

1) All the mice have a Tumor Volume of 45 mm3 on the first timepoint of measurement.This gradually increases or decreases during the treatment period, based on the drug regimen.

2) Out of the four drug regimens specified, Capomulin and Ramicane seem to have smaller mean and median. However, these values for Infubinol and Ceftamine are higher than that of the measure at the first timepoint. This shows that both Capomulin and Ramicane are effective in treating squamous cell carcinoma (SCC). Also, the data shows that Ramicane is slightly better than Capomulin.

3) The correlation coefficient between mouse weight and average tumor volume for the Capomulin regimen is 0.84. This is greater than 0.7 and so indicates that the correlation is strong. The strength of this correlation for Ramicane is 0.81 which shows that both the drug work the similar way.

4) The line plot of tumor volume vs. time point for the mouse r554 treated with Capomulin shows that there is a slight increase in the tumor volumes measured at the timepoints 5, 20,35 and 45. However, these increases are insignificant compared to the decreases at other points. The graph does show the positive effect of the drug in treating SCC.

5) From the raw data, it can be seen that the drugs aren't very effective on younger mice (age 1 or 2 months). Almost all the drugs except Ramicane have showed an increase in the Tumor Volume at the end of the treatment cycle than that at the start. An analysis around the age data would help understand the effectiveness of the drug on different age groups.

```python
# Download the required built-in modules
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Source files to load
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the files into Pandas DataFrames
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study_combined = pd.merge(mouse_metadata,study_results,how = "outer", on = "Mouse ID")
mouse_study_combined
```

```python
# Check for the number of unique mice IDs
mouse_study_combined["Mouse ID"].nunique()
```

```python
# Check for any mouse ID with duplicate time points 
duplicate_mice = mouse_study_combined[mouse_study_combined.duplicated(["Mouse ID","Timepoint"])]
duplicate_mice
```

```python
# Display all data associated with the mouse ID identified above
duplicate_mice_all = mouse_study_combined[mouse_study_combined.duplicated(["Mouse ID","Timepoint"],keep = False)]
duplicate_mice_all
```

```python
# Create a clean DataFrame by dropping the duplicate rows
mouse_study_clean = mouse_study_combined.drop_duplicates(["Mouse ID","Timepoint"])
mouse_study_clean
```

```python
# Check for the number of mice in the clean DataFrame.
mouse_study_clean["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Use the groupby method to generate the mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen.

drug_regimen = mouse_study_clean.groupby("Drug Regimen")

tumor_volume_mean = drug_regimen["Tumor Volume (mm3)"].mean()
tumor_volume_median = drug_regimen["Tumor Volume (mm3)"].median() 
tumor_volume_variance = drug_regimen["Tumor Volume (mm3)"].var()
tumor_volume_standard_deviation = drug_regimen["Tumor Volume (mm3)"].std()
tumor_volume_SEM = drug_regimen["Tumor Volume (mm3)"].sem()

summary_statistics_data = {"Mean (Tumor Volume)": tumor_volume_mean,
                           "Median (Tumor Volume)": tumor_volume_median,
                           "Variance (Tumor Volume)": tumor_volume_variance,
                           "Standard Deviation (Tumor Volume)": tumor_volume_standard_deviation,
                           "SEM (Tumor Volume)": tumor_volume_SEM
                          }

#Combine the five series objects into a single summary statistics DataFrame.
summary_statistics = pd.DataFrame(summary_statistics_data)
summary_statistics
```

```python
# use the agg method to produce the same summary statistics table by using a single line of code.

summary_statistics_agg = mouse_study_clean.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
summary_statistics_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Get the last (greatest) timepoint for each mouse along with the tumor volume
mouse_last_timepoint = mouse_study_clean.groupby("Mouse ID", as_index=False).last()
mouse_last_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_regimen = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume = []

# Calculate Quartiles, IQR, Upper and Lower bounds for the 4 treatment regimen
for i in treatment_regimen:
    
#   Tumor Volume for the 4 Treatment Regimen     
    tumor_data = mouse_last_timepoint.loc[mouse_last_timepoint["Drug Regimen"] == i, "Tumor Volume (mm3)"]
    
#   Mouse and Timepoint data for the 4 Treatment Regimen - Used for printing outliers    
    mouse_data = mouse_last_timepoint.loc[mouse_last_timepoint["Drug Regimen"] == i, "Mouse ID"]
    timepoint_data = mouse_last_timepoint.loc[mouse_last_timepoint["Drug Regimen"] == i, "Timepoint"]

#   Tumor Volume - Used for plotting    
    tumor_volume.append(tumor_data)
        
#   Calculate and Print Quartiles, IQR and Median        
    quartiles = tumor_data.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    print(f"{i}:")
    print(f"The lower quartile is {lowerq}")
    print(f"The upper quartile is {upperq}")
    print(f"The interquartile range is {iqr}")
    print(f"The median is: {quartiles[0.5]} ")

#   Calculate and Print Lower and Upper Bounds    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    
#   Identify and Print Outliers along with the Mouse ID and Timepoint details    
    outlier_tumor = tumor_data.loc[(tumor_data < lower_bound) | (tumor_data > upper_bound)]

    if outlier_tumor.count() > 0:
        print(f"Outliers: {outlier_tumor.count()}")
        for i in outlier_tumor.index:
            print(f"The Tumor Volume(mm3) of {outlier_tumor[i]} for mouse {mouse_data[i]} at the final timepoint {timepoint_data[i]} is an outlier.")
    else:
        print(f"No outliers.")
    print(f"----------------------------------------------------------")  
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest 
fig1, ax1 = plt.subplots()
ax1.set_title("Final Tumor Volume of mice across Drug Regimens")
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.set_xlabel("Drug Regimen")
ax1.boxplot(tumor_volume, flierprops = {"markerfacecolor": "red", "marker": "s"})
ax1.set_xticklabels(treatment_regimen)
plt.show()
```

## Correlation and Regression

```python
# Calculate and print the correlation coefficient for mouse weight and average tumor volume for the Capomulin regimen
correlation,p_value = st.pearsonr(capomulin_regimen["Mouse Weight (g)"],capomulin_regimen["Average Tumor Volume (mm3)"])

print(f"The correlation coefficient between mouse weight and average tumor volume for the Capomulin regimen is {correlation}")
```

```python
# Plot the linear regression model on top of the previous scatter plot.

# Generate the line equation using Mouse weight and average tumor volume
x_values = capomulin_regimen["Mouse Weight (g)"]
y_values = capomulin_regimen["Average Tumor Volume (mm3)"]

slope, intercept, rvalue, pvalue, stderr = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values)
plt.title("Average Tumor Volume vs. Mouse Weight for the Capomulin Regimen")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(18,37.5),fontsize=15,color="red")

plt.show()
```
